## Grid Adventure V1

### Overview

Grid Adventure V1 is, at its core, a grid‑world environment enriched with varied mechanics. The input is a MxN grid (positive integers M, N). Each tile can contain multiple objects (floor, agent, key, door, etc.), each with distinct behavior. The objective is collect any gems if present, before reaching the exit tile.

If you didn’t skip the getting started section at the top of this notebook, you should already be familiar with the game.

### Constraints (Project Scope)

To keep the capstone project tractable, we enforce the following constraints:

- There is exactly 1 agent in the grid, with starting health **at most 5**.
- There is exactly 1 exit tile in the grid.
- The agent moves one tile per step, except under the speed (Boots) power‑up, which doubles movement to two tiles per step.
- Each entity type has a fixed, known cost, damage, or reward (cost = –reward) where applicable:
  - Coins: reward 5.
  - Turn cost: 3 (Every turn incurs a cost of 3 (or reward -3) except for the final turn to win the game).
  - Lava: damage 2.
- Any key can be used to open any door, but each key can only be used once. 
- It is guaranteed that the initial grid will have an **equal number of keys and locked doors**
- It is also guaranteed that any doors would be **locked in the initial state.**
- There are **at most 5** gems in any grid.
- There are **at most 5** key-door pairs in any grid.
- There are **at most 5** pushable boxes in any grid.
- There can be **any number of** coins or lava.
- Power‑up durations are fixed:
  - Boots (Speed): 5 steps.
  - Ghost (Phasing): 5 steps.
  - Shield: 5 uses.
- There are **at most 3** of each powerup type in the grid.
- There will be **no overlapping entities in the initial state** of the grid (i.e each cell contains only `FloorEntity`, or `FloorEntity` with exactly 1 other entity)
- The maximum grid size is in the graded test cases will be **15x15**

````python
import heapq
from collections import deque, namedtuple
from typing import Any

from grid_adventure.grid import GridState, to_state
from grid_adventure.env import ImageObservation
from grid_adventure.step import Action

# Direction deltas for each movement action
DX = {Action.UP: 0, Action.DOWN: 0, Action.LEFT: -1, Action.RIGHT: 1}
DY = {Action.UP: -1, Action.DOWN: 1, Action.LEFT: 0, Action.RIGHT: 0}
MOVES = (Action.UP, Action.DOWN, Action.LEFT, Action.RIGHT)
ALL = (*MOVES, Action.PICK_UP, Action.USE_KEY)

# Compact search state: all fields are plain ints/tuples
# pos: agent (x, y), gems: bitmask, bit i set  <->  gem i collected
# keys: bitmask, bit i set <-> key i in inventory
# doors: bitmask, bit i set <-> door i unlocked
# coll: bitmask: bit i set <-> collectible i already picked up
# boxes: tuple of box positions encoded as x*H + y
# speed/phase/shield: turns remaining on speed/phase/shield power-up
# hp: agent's current health

SS = namedtuple('SS', ['pos', 'gems', 'keys', 'doors', 'boxes', 'speed', 'phase', 'shield', 'hp', 'coll'])
_CLASS_NAMES = ['floor', 'wall', 'human', 'exit', 'coin', 'gem', 'key', 'door', 'lava', 'box', 'boots', 'shield', 'ghost']

def get_model(device: str = "cpu", dtype: str | None = None):
    """
    Return a TorchScript model loaded from an embedded, base64-encoded compressed blob.
    Self-contained: no need for the original Python class.

    Args:
        device: Where to map the model (e.g., "cpu", "cuda", "cuda:0").
        dtype: Optional dtype to convert parameters/buffers to (e.g., "float32", "float16").
    """
    import base64, io, torch
    import zlib as _z; _decomp = _z.decompress
    _blob_b64 = ""
    _raw = _decomp(base64.b64decode(_blob_b64))
    buf = io.BytesIO(_raw)
    m = torch.jit.load(buf, map_location=device)
    if dtype is not None:
        dt = getattr(torch, dtype) if isinstance(dtype, str) else dtype
        for p in m.parameters():
            p.data = p.data.to(dt)
        for b in m.buffers():
            b.data = b.data.to(dt)
    m.eval()
    return m



class Agent:
    TILE = 3       
    COIN = 5      
    MAX_EXP = 100000  # A* node expansion limit to prevent runaway on huge levels

    def __init__(self):
        self.path = [] # remaining actions to execute for current plan
        self.ready = False # True if initialise() called for current level
        self.cur = None  # Track current state SS (in sync with game)
        self.model = get_model()
        self.prev_turn = -1  # Turn counter from last step, used for reset detection

    def _current_turn(self, state):
        # Extract turn number from either observation type
        if isinstance(state, GridState):
            return state.turn
        return state['info']['status']['turn']

    def step(self, state: GridState | ImageObservation) -> Action:
        # Detect level reset: turn dropped back to 0 (or behind prev_turn)
        cur_turn = self._current_turn(state)
        if self.ready and cur_turn < self.prev_turn:
            self.ready = False
            self.path = []
            self.cur = None
        self.prev_turn = cur_turn

        if isinstance(state, GridState):
            s0 = to_state(state)  # convert GridState to ECS State
            if getattr(s0, 'win', False) or getattr(s0, 'lose', False):
                return Action.WAIT
            if not self.ready:
                self.initialise(state)
                self.cur = self.start
        elif isinstance(state, dict) and 'image' in state:
            if state['info']['status']['phase'] != 'ongoing':
                return Action.WAIT
            if not self.ready:
                if self.model is None:
                    return Action.WAIT
                self.initialise(state)
                self.cur = self.start
        else:
            return Action.WAIT

        # If current plan exhausted, replan from current internal state
        if not self.path:
            self.path = self.plan(self.cur)
        if not self.path:
            return Action.WAIT

        # Pop next action, update internal state, return the action
        a = self.path.pop(0)
        ns = self.transition(self.cur, a)
        if ns is not None:
            self.cur = ns
        return a

    def info(self) -> dict:
        return {'plan_len': len(self.path)}

    def parse(self, obs):
        import numpy as np

        # Framework path: ImageObservation with keys 'image' and 'info'
        if isinstance(obs, dict) and 'image' in obs and 'info' in obs:
            return self.parse_image_obs_to_gridstate(obs)

        image_obs = obs.get('image_obs')
        grid_obs = obs.get('grid_obs')

        if image_obs is None or grid_obs is None:
            raise ValueError("parse() expects a dict with keys 'image_obs' and 'grid_obs'")

        WIDTH, HEIGHT = grid_obs.width, grid_obs.height

        # Build ground-truth grid from GridState
        def gs_label(gs, x, y):
            for obj in gs.objects_at((x, y)):
                n = type(obj).__name__.lower()
                if 'agent' in n: return 'human'
                if 'wall' in n: return 'wall'
                if 'exit' in n: return 'exit'
                if 'lava' in n: return 'lava'
                if 'gem' in n: return 'gem'
                if 'unlock' in n: return 'opened'
                if 'locked' in n: return 'locked'
                if 'key' in n: return 'key'
                if 'coin' in n: return 'coin'
                if 'box' in n: return 'box'
                if 'speed' in n or 'boot'  in n: return 'boots'
                if 'phas' in n or 'ghost' in n: return 'ghost'
                if 'shield' in n: return 'shield'
            return 'floor'

        actual = [[gs_label(grid_obs, x, y) for x in range(WIDTH)] for y in range(HEIGHT)]
        cnn = self.parse_image(image_obs['image'], WIDTH, HEIGHT)
        cnn_by_row = [[cnn[x][y] for x in range(WIDTH)] for y in range(HEIGHT)]

        # Find mismatches
        mismatches = []
        for y in range(HEIGHT):
            for x in range(WIDTH):
                if actual[y][x] != cnn[x][y]:
                    mismatches.append((x, y, actual[y][x], cnn[x][y]))

        col_w = 10
        header = f"{'ACTUAL (GridState)':<{col_w * WIDTH}}  {'CNN PREDICTION'}"
        print(header)
        print('-' * len(header))
        for y in range(HEIGHT):
            left  = ''.join(f'{actual[y][x]:<{col_w}}' for x in range(WIDTH))
            right = ''.join(f'{cnn[x][y]:<{col_w}}' for x in range(WIDTH))
            print(f'{left}  {right}')

        print()
        if mismatches:
            print(f'MISMATCHES ({len(mismatches)}):')
            for x, y, truth, pred in mismatches:
                print(f'  ({x},{y}): actual={truth!r:10s} cnn={pred!r}')
        else:
            print('Perfect match: CNN agrees with GridState on all tiles')

    def parse_image_obs_to_gridstate(self, obs):
        """
        Convert an ImageObservation (keys 'image', 'info') to GridState using CNN classifier
        """
        from grid_adventure.entities import (
            AgentEntity, FloorEntity, WallEntity, ExitEntity, CoinEntity,
            GemEntity, KeyEntity, LockedDoorEntity,
            BoxEntity, LavaEntity, SpeedPowerUpEntity, ShieldPowerUpEntity,
            PhasingPowerUpEntity,
        )
        from grid_adventure.movements import MOVEMENTS
        from grid_adventure.objectives import OBJECTIVES

        info = obs['info']
        config = info['config']
        status = info['status']

        WIDTH = config['width']
        HEIGHT = config['height']

        # Resolve movement and objective objects by name
        movement = MOVEMENTS.get(config['movement'])
        obj_by_name = {o.name: o for o in OBJECTIVES.values()}
        objective = obj_by_name.get(config['objective'])

        # Fall back to first available if names don't match
        if movement is None:
            movement = next(iter(MOVEMENTS.values()))
        if objective is None:
            objective = next(iter(OBJECTIVES.values()))

        seed = config['seed'] if config['seed'] != -1 else None
        turn_limit = config['turn_limit'] if config['turn_limit'] != -1 else None

        gs = GridState(
            width=WIDTH, height=HEIGHT,
            movement=movement, objective=objective,
            seed=seed,
            turn=status['turn'],
            score=status['score'],
            turn_limit=turn_limit,
            win=(status['phase'] == 'win'),
            lose=(status['phase'] == 'lose'),
        )

        # Fill every cell with a floor tile first
        for x in range(WIDTH):
            for y in range(HEIGHT):
                gs.add((x, y), FloorEntity())

        if self.model is None:
            return gs

        # Classify tiles with CNN and add entities on top
        grid = self.parse_image(obs['image'], WIDTH, HEIGHT)
        _ENTITY_MAP = {
            'human': AgentEntity,
            'wall': WallEntity,
            'exit': ExitEntity,
            'lava': LavaEntity,
            'gem': GemEntity,
            'door': LockedDoorEntity,
            'key': KeyEntity,
            'coin': CoinEntity,
            'box': BoxEntity,
            'boots': SpeedPowerUpEntity,
            'ghost': PhasingPowerUpEntity,
            'shield': ShieldPowerUpEntity,
        }
        for x in range(WIDTH):
            for y in range(HEIGHT):
                cls = _ENTITY_MAP.get(grid[x][y])
                if cls is not None:
                    gs.add((x, y), cls())

        return gs

    # Look up entity's (x, y) position from ECS state
    def get_pos(self, s0, eid):
        p = s0.position.get(eid)
        return (p.x, p.y) if p else None

    def parse_image(self, image, W, H):
        """
        Slice RGBA grid image into W*H individual tile crops, classify each with CNN model
        return 2D list: entity_grid[x][y] = class name string (from _CLASS_NAMES)
        Uses same RGBA normalisation as training (mean/std per channel)
        """
        import numpy as np
        import torch
        from PIL import Image as PILImage

        ph, pw = image.shape[0], image.shape[1]
        tile_height = ph / H
        tile_width = pw / W

        tiles, coords = [], []
        for y in range(H):
            for x in range(W):
                # Round boundaries to nearest pixel
                y0, y1 = round(y * tile_height), round((y + 1) * tile_height)
                x0, x1 = round(x * tile_width), round((x + 1) * tile_width)
                crop = image[y0:y1, x0:x1]

                # Resize to 64x64 and scale to [0,1]
                pil = PILImage.fromarray(crop, mode='RGBA').resize((64, 64), PILImage.BILINEAR)
                t = torch.tensor(np.array(pil), dtype=torch.float32).permute(2, 0, 1) / 255.0
                tiles.append(t)
                coords.append((x, y))

        batch = torch.stack(tiles) # (N, 4, 64, 64)
        with torch.no_grad():
            preds = self.model(batch).argmax(1) # (N,) class indices

        grid = [[None] * H for _ in range(W)]
        for (x, y), class_idx in zip(coords, preds.tolist()):
            grid[x][y] = _CLASS_NAMES[class_idx]
        return grid

    def initialise(self, state: 'GridState | ImageObservation'):
        """
        Parse level once, build all static data structures needed for search
        Accepts either a GridState (ECS extraction) or an ImageObservation (CNN classification)
        Called only once on first step() call after each level reset
        """
        walls, lava = set(), set()
        gem_pos, key_pos, door_pos, box_pos = [], [], [], []
        coll_pos, coll_type = [], []
        exit_pos, agent_pos = None, None
        hp = 5
        speed, phase, shield = 0, 0, 0
        keys_held = 0

        if isinstance(state, GridState):
            s0 = to_state(state)
            self.W, self.H = state.width, state.height

            # Walls: ECS blocking store, excluding pushable boxes and locked doors
            non_wall_blocking = set(s0.pushable) | set(s0.locked)
            g = self.get_pos
            walls = {
                self.get_pos(s0, eid)
                for eid in s0.blocking
                if eid not in non_wall_blocking and self.get_pos(s0, eid) is not None
            }

            agent_id = next(iter(s0.agent))
            agent_pos = g(s0, agent_id)
            hp_obj = s0.health.get(agent_id)
            hp = getattr(hp_obj, 'current_health', 5) if hp_obj else 5

            gem_pos = [pos for e in s0.requirable if (pos := g(s0, e)) is not None]
            key_pos = [pos for e in s0.key if (pos := g(s0, e)) is not None]
            door_pos = [pos for e in s0.locked if (pos := g(s0, e)) is not None]
            box_pos = [pos for e in s0.pushable   if (pos := g(s0, e)) is not None]
            lava = {g(s0, e) for e in s0.damage if g(s0, e)}
            exit_pos = g(s0, next(iter(s0.exit)))

            gem_set, key_set = set(s0.requirable), set(s0.key)
            for eid in s0.collectible:
                if eid in gem_set or eid in key_set:
                    continue
                pos = g(s0, eid)
                if not pos:
                    continue
                coll_t = ('coin'  if eid in s0.rewardable else
                          'speed' if eid in s0.speed else
                          'phase' if eid in s0.phasing else 'shield')
                coll_pos.append(pos)
                coll_type.append(coll_t)

        else:  # ImageObservation
            info = state['info']
            self.W = info['config']['width']
            self.H = info['config']['height']
            grid = self.parse_image(state['image'], self.W, self.H)

            for x in range(self.W):
                for y in range(self.H):
                    entity = grid[x][y]
                    if entity == 'wall': walls.add((x, y))
                    elif entity == 'lava': lava.add((x, y))
                    elif entity == 'gem': gem_pos.append((x, y))
                    elif entity == 'key': key_pos.append((x, y))
                    elif entity == 'door': door_pos.append((x, y))
                    elif entity == 'box': box_pos.append((x, y))
                    elif entity == 'exit': exit_pos = (x, y)
                    elif entity == 'human': agent_pos = (x, y)
                    elif entity == 'coin': coll_pos.append((x, y)); coll_type.append('coin')
                    elif entity == 'boots': coll_pos.append((x, y)); coll_type.append('speed')
                    elif entity == 'ghost': coll_pos.append((x, y)); coll_type.append('phase')
                    elif entity == 'shield': coll_pos.append((x, y)); coll_type.append('shield')

            agent_info = info['agent']
            hp_info = agent_info.get('health') or {}
            hp = hp_info.get('current_health', 5) if isinstance(hp_info, dict) else 5
            keys_held = sum(1 for item in agent_info['inventory'] if item['type'] == 'key')
            for eff in agent_info['effects']:
                amount = eff['limit_amount']
                if eff['type'] == 'PHASING': phase = max(phase, amount)
                elif eff['type'] == 'SPEED': speed = max(speed, amount)
                elif eff['type'] == 'IMMUNITY': shield = max(shield, amount)

        # Shared structure-building (identical for both input types)
        self.walls = frozenset(walls)
        self.lava = frozenset(lava)
        self.exit = exit_pos
        self.gem_pos = gem_pos
        self.door_pos = door_pos
        self.box_pos = box_pos
        self.coll_pos = coll_pos
        self.coll_type = coll_type

        # Keys: virtual inventory slots prepended so bit indices stay consistent
        self.key_pos = [(-1, -1)] * keys_held + key_pos

        # Reverse-lookup dicts: position -> entity index
        self.gem_at = {p: i for i, p in enumerate(gem_pos) if p}
        self.key_at = {p: i + keys_held for i, p in enumerate(key_pos)}
        self.door_at = {p: i for i, p in enumerate(door_pos) if p}
        self.coll_at = {p: i for i, p in enumerate(coll_pos)}

        # Precompute BFS distance maps from each gem and from exit
        self.dist = {}
        self.dist_phase = {}  # wall-ignoring BFS for phasing heuristic
        for gp in gem_pos:
            self.dist[gp] = self.bfs(gp)
            self.dist_phase[gp] = self.bfs_phase(gp)
        if exit_pos:
            self.dist['exit'] = self.bfs(exit_pos)
            self.dist_phase['exit'] = self.bfs_phase(exit_pos)

        self.goal_gems = (1 << len(gem_pos)) - 1

        if agent_pos is None:
            return  # Cannot locate agent in level; self.ready stays False
        self.start = SS(
            pos=agent_pos, gems=0,
            keys=(1 << keys_held) - 1, doors=0,
            boxes=tuple(p[0] * self.H + p[1] for p in box_pos),
            speed=speed, phase=phase, shield=shield, hp=hp, coll=0,
        )
        self.ready = True

    def bfs(self, src):
        """
        BFS from src on static grid (walls blocked, doors/lava ignored)
        Returns dict {(x,y): num_steps_from_src}
        Used as admissible lower bound on movement distance for heuristic
        """
        d = {src: 0}
        q = deque([src])
        while q:
            x, y = q.popleft()
            for nx, ny in ((x-1,y),(x+1,y),(x,y-1),(x,y+1)):
                if (0 <= nx < self.W and 0 <= ny < self.H and (nx,ny) not in self.walls and (nx,ny) not in d):
                    d[(nx,ny)] = d[(x,y)] + 1
                    q.append((nx,ny))
        return d

    def bfs_phase(self, src):
        """
        BFS from src ignoring walls (simulate phasing permanently active)
        Returns dict {(x,y): num_steps_from_src}
        Used by heuristic when ss.phase > 0 to give tighter distance estimates through walls that phasing can bypass
        """
        d = {src: 0}
        q = deque([src])
        while q:
            x, y = q.popleft()
            for nx, ny in ((x-1,y),(x+1,y),(x,y-1),(x,y+1)):
                if 0 <= nx < self.W and 0 <= ny < self.H and (nx,ny) not in d:
                    d[(nx,ny)] = d[(x,y)] + 1
                    q.append((nx,ny))
        return d

    def heuristic(self, ss):
        """
        Weighted heuristic for A*

        If all gems collected: estimate cost = BFS_dist(pos -> exit) * MULT
        Else: for each uncollected gem, estimate = (dist_to_gem + dist_gem_to_exit) * MULT, take maximum
        Extra TILE per uncollected gem accounts for PICK_UP action

        When ss.phase > 0, uses wall-ignoring BFS distances (dist_phase) for a tighter estimate
        """
        MULT = 1.5
        dist = self.dist_phase if ss.phase > 0 else self.dist
        uncollected_gems_positions = [self.gem_pos[i] for i in range(len(self.gem_pos)) if not (ss.gems >> i & 1)]
        if not uncollected_gems_positions:
            return dist.get('exit', {}).get(ss.pos, 999) * MULT

        ex = dist.get('exit', {})
        hv = 0
        for gem in uncollected_gems_positions:
            dt = dist.get(gem, {}).get(ss.pos, 999) # dt = current pos -> gem
            de = ex.get(gem, 999) # de = gem -> exit
            hv = max(hv, (dt + de) * MULT)
        return hv + len(uncollected_gems_positions) * self.TILE

    def one_step(self, ss, dx, dy):
        """
        Attempt single-tile move in direction (dx, dy)
        Returns new SS on success, or None if move is illegal (out of bounds, blocked, lethal)

        - Boundary checks
        - Wall blocking (bypassed when phasing)
        - Locked door blocking (bypassed when phasing)
        - Box pushing: box must have clear cell to slide into
        - Lava damage: shield -= 1, else hp -= 2, lethal if hp <= 0 (bypassed when phasing)
        """
        x, y = ss.pos
        new_x, new_y = x + dx, y + dy
        if not (0 <= new_x < self.W and 0 <= new_y < self.H): # Out of bounds check
            return None
        phase_left = ss.phase > 0  # no. of turns left for phase power-up
        if (new_x, new_y) in self.walls and not phase_left: # Wall blocking, phase NOT active
            return None
        di = self.door_at.get((new_x, new_y))
        if di is not None and not (ss.doors >> di & 1) and not phase_left: # Locked door blocking, phase NOT active
            return None

        # Box pushing: if a box occupies destination, push it one step further
        boxes = list(ss.boxes)
        box_pos = new_x * self.H + new_y
        box_idx = next((i for i, b in enumerate(boxes) if b == box_pos), None)
        if box_idx is not None:
            bx2, by2 = new_x + dx, new_y + dy
            if not (0 <= bx2 < self.W and 0 <= by2 < self.H): # New box position out of bounds
                return None
            if (bx2, by2) in self.walls: # Wall blocking
                return None
            if any(b == bx2 * self.H + by2 for b in boxes): # Another box occupies push destination
                return None   
            di2 = self.door_at.get((bx2, by2))
            if di2 is not None and not (ss.doors >> di2 & 1): # Push destination has locked door
                return None   
            boxes[box_idx] = bx2 * self.H + by2 # Update box_pos for boxes[box_idx]

        # Lava: apply damage unless phasing active
        hp, shd = ss.hp, ss.shield
        if (new_x, new_y) in self.lava and not phase_left:
            if shd > 0:
                shd -= 1 # shield absorbs hit
            else:
                hp -= 2
                if hp <= 0: return None # agent dies, discard this branch

        return ss._replace(pos=(new_x, new_y), boxes=tuple(boxes), hp=hp, shield=shd)

    def move(self, ss, dx, dy):
        """
        Execute a movement action when speed power-up active

        Always decrements phase by 1
        With speed active: attempts 2nd tile in SAME direction
        If 2nd step blocked, agent stops on first tile
        """
        s1 = self.one_step(ss, dx, dy)
        if s1 is None:
            return None
        s1 = s1._replace(phase=max(0, ss.phase - 1))
        if ss.speed > 0:
            s1 = s1._replace(speed=ss.speed - 1)
            s2 = self.one_step(s1, dx, dy) # attempt second step
            return s2 if s2 else s1 # stop on first tile if blocked
        return s1

    def pickup(self, ss):
        """
        PICK_UP action at agent's current position
        Priority: gem > key > other collectibles (coin / power-up)
        Returns None if nothing at current position can be picked up
        Power-ups grant 5 turns of effect
        """
        x, y = ss.pos
        gem_idx = self.gem_at.get((x, y))
        if gem_idx is not None and not (ss.gems >> gem_idx & 1):
            return ss._replace(gems=ss.gems | (1 << gem_idx))
        key_idx = self.key_at.get((x, y))
        if key_idx is not None and not (ss.keys >> key_idx & 1):
            return ss._replace(keys=ss.keys | (1 << key_idx))
        coll_idx = self.coll_at.get((x, y))
        if coll_idx is not None and not (ss.coll >> coll_idx & 1):
            coll_t = self.coll_type[coll_idx]
            # speed/phase reset to 5 (time-based, doesn't stack duration)
            # shield stacks by usage: add 5 to existing shield count
            extra = ({'speed':  5} if coll_t == 'speed' else
                     {'phase':  5} if coll_t == 'phase' else
                     {'shield': ss.shield + 5} if coll_t == 'shield' else {})
            return ss._replace(coll=ss.coll | (1 << coll_idx), **extra)
        return None

    def use_key(self, ss):
        """
        USE_KEY action: unlock min(spare_keys, surrounding_locked_doors) doors at once
        Checks current cell first, then LEFT, RIGHT, UP, DOWN
        Requires at least one spare key
        """
        spare_keys = bin(ss.keys).count('1') - bin(ss.doors).count('1')
        if spare_keys <= 0:
            return None
        x, y = ss.pos
        # Priority: current cell, then LEFT, RIGHT, UP, DOWN
        candidates = [(0,0), (-1,0), (1,0), (0,-1), (0,1)]
        new_doors = ss.doors
        used = 0
        for dx, dy in candidates:
            if used >= spare_keys:
                break
            door_idx = self.door_at.get((x + dx, y + dy))
            if door_idx is not None and not (new_doors >> door_idx & 1):
                new_doors |= (1 << door_idx)
                used += 1
        if used == 0:
            return None
        return ss._replace(doors=new_doors)

    def transition(self, ss, a):
        """
        Dispatch action and return new SS, or None if illegal
        For non-move actions (PICK_UP / USE_KEY), game still decrements power-up timers by 1 (phase, speed)
        """
        if a in MOVES:
            return self.move(ss, DX[a], DY[a])
        if a == Action.PICK_UP:
            new_state = self.pickup(ss)
        elif a == Action.USE_KEY:
            new_state = self.use_key(ss)
        else:
            return None
        if new_state is None:
            return None
        # Decrement power-up timers
        return new_state._replace(phase=max(0, new_state.phase - 1), speed=max(0, new_state.speed - 1))

    def transition_cost(self, ss, ns, a):
        """
        Edge cost for A* g-value:
          - Reach goal: 0
          - Pick up coin: TILE - COIN = 3 - 5 = -2
          - Everything else: TILE = 3
        """
        if ns.gems == self.goal_gems and ns.pos == self.exit:
            return 0 # goal achieved, reach exit
        if a == Action.PICK_UP:
            # Detect which coll bits newly set by this action
            new_bits = (ns.coll ^ ss.coll) & ns.coll
            for i in range(len(self.coll_type)):
                if (new_bits >> i & 1) and self.coll_type[i] == 'coin':
                    return self.TILE - self.COIN  # coin pickup
        return self.TILE

    def plan(self, ss0):
        """
        Weighted A* search from ss0 to goal state

        Goal: ss.gems == goal_gems (all gems collected) AND ss.pos == exit
        Heap stores (f, tie, g, ss), where f = g + h(ss)
        'tie': monotone counter that breaks ties without comparing SS objects
        'best': tracks lowest g seen per state to skip stale heap entries
        Expansion capped at MAX_EXP: returns partial path to best state if hit
        """
        if not self.exit: return []
        tie = 0
        h0 = self.heuristic(ss0)
        heap = [(h0, tie, 0, ss0)]
        best = {ss0: 0}
        path = {} # path[new_state] = (prev_state, action) for path reconstruction
        exp = 0
        best_partial = None # best non-start state found (lowest heuristic)
        best_partial_h = h0
        while heap:
            _, _, g, ss = heapq.heappop(heap)
            # Skip stale entries: cheaper path to ss already expanded
            if g > best.get(ss, float('inf')):
                continue
            if ss.gems == self.goal_gems and ss.pos == self.exit:
                return self.reconstruct(path, ss, ss0) # goal state reached
            # Track best partial state for fallback when expansion limit hit
            if ss != ss0 and ss in path:
                h = self.heuristic(ss)
                if h < best_partial_h:
                    best_partial_h = h
                    best_partial = ss
            if exp >= self.MAX_EXP:
                break # expansion limit hit
            exp += 1
            for a in ALL:
                new_state = self.transition(ss, a)
                if new_state is None or new_state == ss:
                    continue # illegal action or no state change
                new_g = g + self.transition_cost(ss, new_state, a)
                if new_g >= best.get(new_state, float('inf')):
                    continue # already found cheaper path to new_state
                best[new_state] = new_g
                path[new_state] = (ss, a)
                tie += 1
                heapq.heappush(heap, (new_g + self.heuristic(new_state), tie, new_g, new_state))
        # Return partial path to best state if expansion limit exceeded
        if best_partial is not None:
            return self.reconstruct(path, best_partial, ss0)
        return [] # no path found within expansion limit

    def reconstruct(self, path, goal, start):
        """
        Reconstruct sequence of actions using path map backward from goal to start
        """
        acts = []
        ss = goal
        while ss != start:
            ss, a = path[ss]
            acts.append(a)
        return list(reversed(acts))
````

````python
import os
import re
import sys
import subprocess
import textwrap

# Compare these agent implementations (module_name, source_file)
AGENT_IMPLS = [
    ("agent", "agent.py"),
    ("agent_2", "agent_2.py"),
    ("agent_3", "agent_3.py"),
    ("agent_4", "agent_4.py"),
]

UPLOAD_LIMIT_BYTES = 2 * 1024 * 1024  # 2 MB
MEMORY_LIMIT_BYTES = int(1.5 * 1024 * 1024 * 1024)  # 1.5 GB
SEEDS = [0, 42]

# Use one built-in level and one hard custom level (if defined) for a more realistic peak.
TEST_CASES = [
    ("combined_mechanics", "grid_adventure.levels.intro", "build_level_combined_mechanics", -84, -126, 10),
]
if "CUSTOM_LEVELS" in globals() and CUSTOM_LEVELS:
    custom_name, _, custom_max, custom_min = CUSTOM_LEVELS[0]
    TEST_CASES.append((custom_name, "__main__", "CUSTOM_LEVELS[0][1]", custom_max, custom_min, 10))


def _format_bytes(num_bytes: int) -> str:
    return f"{num_bytes / (1024 * 1024):.3f} MB"


def _extract_peak_rss_bytes(stderr_text: str):
    # macOS /usr/bin/time -l prints: "<num>  maximum resident set size"
    m = re.search(r"(\d+)\s+maximum resident set size", stderr_text)
    return int(m.group(1)) if m else None


def _run_memory_probe(module_name: str, builder_module: str, builder_expr: str, max_reward: int, min_reward: int, seed: int, time_limit: int):
    code = f"""
from importlib import import_module
from utils import evaluate_level

Agent = import_module({module_name!r}).Agent
if {builder_module!r} == "__main__":
    builder = eval({builder_expr!r})
else:
    builder = getattr(import_module({builder_module!r}), {builder_expr!r})

result = evaluate_level(
    Agent,
    builder,
    observation_type='gridstate',
    max_total_reward={max_reward},
    min_total_reward={min_reward},
    turn_limit=150,
    time_limit={time_limit},
    seed={seed},
)
print(result['performance'])
print(result['timeout'])
print(result['runtime (sec)'])
print(result['total_reward'])
print(result['win'])
"""

    cmd = ["/usr/bin/time", "-l", sys.executable, "-c", textwrap.dedent(code)]
    proc = subprocess.run(cmd, capture_output=True, text=True)

    peak_rss = _extract_peak_rss_bytes(proc.stderr)
    out_lines = [line.strip() for line in proc.stdout.splitlines() if line.strip()]

    parsed = {
        "return_code": proc.returncode,
        "peak_rss_bytes": peak_rss,
        "performance": float(out_lines[0]) if len(out_lines) > 0 else None,
        "timeout": out_lines[1] == "True" if len(out_lines) > 1 else None,
        "runtime_sec": float(out_lines[2]) if len(out_lines) > 2 else None,
        "total_reward": float(out_lines[3]) if len(out_lines) > 3 else None,
        "win": out_lines[4] == "True" if len(out_lines) > 4 else None,
        "stderr_tail": "\n".join(proc.stderr.splitlines()[-8:]),
    }
    return parsed


print("=== Upload Size Check (2 MB limit) ===")
for module_name, filename in AGENT_IMPLS:
    if not os.path.exists(filename):
        print(f"{filename:<12} MISSING")
        continue
    size_bytes = os.path.getsize(filename)
    status = "OK" if size_bytes <= UPLOAD_LIMIT_BYTES else "TOO LARGE"
    print(f"{filename:<12} {size_bytes:>10} bytes  {_format_bytes(size_bytes):>10}  {status}")

if os.path.exists("capstone-project.ipynb"):
    nb_size = os.path.getsize("capstone-project.ipynb")
    print(f"{'capstone-project.ipynb':<12} {nb_size:>10} bytes  {_format_bytes(nb_size):>10}  (not usually submitted as raw file)")

print("\n=== Peak Memory Check (1.5 GB limit) ===")
print(f"Memory limit: {MEMORY_LIMIT_BYTES} bytes ({_format_bytes(MEMORY_LIMIT_BYTES)})")

for module_name, filename in AGENT_IMPLS:
    if not os.path.exists(filename):
        continue

    print(f"\n[{module_name}]")
    worst_peak = None

    for case_name, builder_module, builder_expr, max_reward, min_reward, time_limit in TEST_CASES:
        for seed in SEEDS:
            result = _run_memory_probe(
                module_name=module_name,
                builder_module=builder_module,
                builder_expr=builder_expr,
                max_reward=max_reward,
                min_reward=min_reward,
                seed=seed,
                time_limit=time_limit,
            )

            peak = result["peak_rss_bytes"]
            if peak is not None:
                worst_peak = peak if worst_peak is None else max(worst_peak, peak)

            peak_str = "N/A" if peak is None else f"{peak} ({_format_bytes(peak)})"
            print(
                f"  case={case_name:<18} seed={seed:<3} "
                f"peak_rss={peak_str:<24} perf={result['performance']} "
                f"win={result['win']} timeout={result['timeout']} runtime={result['runtime_sec']}"
            )

            if result["return_code"] != 0:
                print("    subprocess failed; stderr tail:")
                print(textwrap.indent(result["stderr_tail"], prefix="      "))

    if worst_peak is None:
        print("  worst_peak_rss: N/A (could not parse /usr/bin/time output)")
    else:
        status = "OK" if worst_peak <= MEMORY_LIMIT_BYTES else "EXCEEDS LIMIT"
        print(f"  worst_peak_rss: {worst_peak} bytes ({_format_bytes(worst_peak)}) -> {status}")

print("\nTip: Run this cell a few times and keep the worst observed memory value.")
````

````python
def train(net: nn.Module, X: np.ndarray, y: np.ndarray):
    X = torch.tensor(X, dtype=torch.float32)
    y = torch.tensor(y[:, None], dtype=torch.float32)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(net.parameters(), lr=0.01)
    for epoch in range(100):
        optimizer.zero_grad()
        output = net(X)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()

train(net, X, y)
````
